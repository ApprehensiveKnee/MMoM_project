## <span style="color:coral">0</span>. <b>Introduction</b>

<span style="font-size: 1.1em;">
The adsorption of gases on carbon nanofibers (CNFs) gained significant interest, especially in the first years of the new century, due to its wide-ranging applications in fields such as <strong>environmental remediation</strong>, <strong>gas storage</strong>, and <strong>catalysis</strong>. Carbon nanofibers, characterized by their high surface area, tunable porosity, and unique electronic properties, present an ideal material for gas adsorption processes.

Understanding the adsorption mechanisms and capacities of gases on CNFs is crucial for optimizing these processes and enhancing the performance of related technologies.
<br>

<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>: <u>Absorption types: physisorption and chemisorption</u> <br>
Physisorption and chemisorption are two types of physical adsorption, which refer to extracting a substance from a medium by interacting with solid surfaces:
<ul>
    <li><b>Physisorption</b> is the adsorption in which the molecular interactions between the adsorbate molecules and the adsorbent are primarily <b>governed by van der Waals forces</b>;
    <li><b>Chemisorption</b> is the adsorption in which the molecular interactions between the adsorbate molecules and the adsorbent are primarily governed by <b>chemical bonds</b>.
</ul>
In this study, we will <b>focus specifically on the physisorption of gases on CNFs</b>.
In general, Physisorption is faster and more effective than chemisorption. Additionally, Physisorption is more selective, meaning that it can selectively adsorb certain substances over others.
None the less, physisorption is non-specific in nature and is generally regarded as a weak, reversible process governed by competitive adsorption and desorption</div>

</span>

<span style="font-size: 1.1em;">
In this study, we will always consider <b>platelet GNFs</b>: the pores are formed between the graphite platelets, and the adsorption sites are located on the platelet surfaces. The areas of the platelets, as well as the distance between them, can be tuned to optimize the adsorption capacity and selectivity of the CNFs.
</span>

<span style="font-size: 1.1em;">
<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>: <u>GNF structures</u> <br>
GNFs consist of parallel stacks of graphite platelets that are oriented parallel, perpendicular, or at an angle with respect to the fiber axis. These structures are termed <b>tubular</b>, <b>platelet</b>, and <b>herringbone</b>, respectively
</div>
</span>

Regarding hydrogen storage in carbon nanoplatelets, <b>experimental evidence of significant amounts of hydrogen storage has been difficult to reproduce and therefore controversial</b>. Storage is also usually found to be <b>irreversible</b>. One experimental issue is that even if a large amount of storage is possible, there may be <u>kinetic barriers to the storage</u>. ## <span style="color:coral">1</span>. <b>DFT analysis</b>

### <span style="color:coral">1.1</span>. <b>About the typical strenght of the interaction...</b>
<span style="font-size: 1.1em;">
As highlighted by Morris et al. (2007) <a href="#r1">[1]</a>, typically, the binding energy must be on the order of <b>25 kJ/mol (0.25 eV/molecule)</b> or higher to have significant storage at operating temperatures and pressures, while not irreversibly binding the hydrogen. The typical sheet spacing for carbon layers is 3.35 Å, which does not allow for significant hydrogen storage, in spite of the hydrogen weak attraction to the carbon surface. 
Nonetheless, the absorption/storage can be significantly enhanced by increasing the interlayer spacing of the graphite platelets.
<br>

In this regard, one method - through which the GNFs considered in this study are generated<a href="#r2">[2]</a> - consists in <b>catalyzed decomposition of carbon-containing gases and their mixtures over selected metals and alloys at temperatures between 450 and 750°C</b>. IN the various GNFs conformations, the layers are separated at distances that are dependent upon the nature of the catalyst and the gas phase, as well as the reaction conditions, where the minimum value possible is that of single-crystal graphite, 3.35 Å.
</span>

### <span style="color:coral">1.2</span>. <b>QE espresso simulations</b>
To study the absorption phenomena in terms of absorption energies and geometries, the <a href="https://www.quantum-espresso.org/"><b>Quantum Espresso (QE)</b></a> package was used:
<br>
a set of relaxation-oriented scripts has been developed, one for each <em>phase</em> considered for the system (i.e. <b>only</b> graphene, <b>only</b> H<sub>2</sub>, then graphene <em>immersed</em> in a progressively large number of molecules). This was done to study the evolution of the system during absorption, ensuring continuity and consistency in the results.
After each simulation ended, the data was then extracted and examinated.

<span style="color:#FFC300; font-size: 1.4em;"><b>SIMULATIONS DETAILS:</b></span>
- <b><u>ultrasoft pseudopotentials</u></b> were used for the calculations (both for C and H atoms);
- supercells consisting of three graphene layers - <b><u> ABA stacking</b></u> and three layers of vacuum have been used for the absorption simulations;
- a third layer of graphene was added to aid the convergence of the simulations, as explained in <a href="#r1">[1]</a>;
- almost all of the simulations have been carried out in <code>vc-relax</code> mode;
- the <b><u>kinetic energy cut-off</b></u> was set to 40 Ry;
- the <b><u>charge density cut-off</b></u> was set to 400 Ry;
- the <b><u>exchange-correlation functional</b></u> used was LDA;
- <b><u> gaussian smearing</b></u> was used for the electronic occupations (<code>degauss=0.01</code>);
- hydrogen molecules have been added in the simulation with their molecular axis both parallel, perdendicular, and at an angle with respect to the fiber axis. As highlighted in <a href="#r1">[1]</a>, the relative orientation of the hydrogen molecules with respect to the fiber axis proved to <b>NOT</b> be a decisive factor in the study of the absorption phenomena;
- <span style="font-size:1.3em;"> <b><u>both gamma(1 1 1) and k(6 6 2)-point calculations were employed</b></u></span>;

<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>:
Regarding the last point: the gamma point calculations were used primarly in the first phases of the project, to ensure the correctness of the simulation parameters. Since those first simulations were run on a remote PC (2 MPI procs), to add an ulterior aid to the convergence, a constrained optimization was performed: specifically, the movement of the platelets was allowed only in the z-direction, which means that the planes of graphene could not "slide one over" the other. </br>
The k-point calculations were instead performed on a cluster, with up to 2 nodes of 48 MPI procs each, and were run without any kind of constraint on the movement of the atoms. Increased computational resources allowed for a more accurate representation of the system, and the results obtained were more reliable but still quite far from the ones presented in <a href="#r1">[1]</a>
</div>

Both the <b>energy absorption</b> results (incorporating strain energy of the carbon structure) from the gamma and k-point calculations highlight a <b>"more difficult" - non favourable - initial absorption of hydrogen into the carbon platelets</b>, which is then followed by a more favourable "phase" for the remaining molecules. </br>

Regarding the <b>interlayer expansion</b>, the results hereby presented deviate quite significantly from the ones presented in <a href="#r1">[1]</a>. Yet, the trend differs significantly, being the ascent of the curve much more gradual in the results presented here. </br>
This may be done to several factors, such as the different computational resources (k-points) used, different system setup, such as dimension of the supercell and number of elemental atoms, or again different constraints applied to the system at optimization time.

None the less, these calculations indicate an <b>increased possibility for hydrogen storage if the graphene interlayer spacing is allowed to vary and is preexpanded</b>, thus motivating the study of the hydrogen absorption properties at finite temperature and pressure by using GCMC simulations at various fixed graph- ite interlayer spacings.

## <span style="color:coral">2. </span><b>MD and Monte Carlo simulations</b>

<span style="font-size: 1.1em;">
We continue out study considering the <b><u>absorption of H<sub>2</sub> molecules</u></b> on carbon platelets. We simulate the absorption process using a <b>Gran Canonical Monte Carlo (GCMC) ensemble</b> study excess weight absorption and density profiles that will be compared to the theoretical and experimental data. For the simulation process, we use the <a href="https://www.lammps.org/#gsc.tab=0"><b>LAMMPS</b></a> code, while the visualization of the results is carried out using the <a href="https://www.ks.uiuc.edu/Research/vmd/"><b>VMD</b></a> software.
<i>
We study the physoption phenomena of H<sub>2</sub>, trying to find a good parametrization for the chosen potentials that may provide a 'good' agreement with the experimental data, also considering the controversiality (<a href="#r1">[1]</a>, <a href="#r2">[2]</a>, <a href="#r3">[3]</a>) of such data.
</i>

<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>:
The objective of a Monte Carlo (MC) simulation is to generate an ensemble of representative configurations under specific thermodynamics conditions for a complex macromolecular system. Applying random perturbations to the system generates these configurations, employing a Markov chain procedure, compliant with appropriate Boltzmann distribution for the system under study. To properly sample the representative space, the perturbations must be sufficiently large, energetically feasible and highly probable. Monte Carlo simulations do not provide information about time evolution. Rather, they provide an ensemble of representative configurations, and, consequently, conformations from which probabilities and relevant thermodynamic observables, such as the free energy, may be calculated
</div>

### <span style="color:coral">2.1</span>. <b>System setting</b>

<span style="font-size: 1.1em;">
The considered carbon platelets have the following characteristics:

- The graphite platelet is a <u>planar hexagonal lattice of carbon sites</u>, held in place throughout the simulation.
Diffently from <a href="#r3">[3]</a>, where che carbon plates are considered ridigly, in this study the carbon atoms are allowed to move around their positions, constrained by an elastic potential, used to keep the atoms and platelets in place.
- The distance between the platelets is (approximately) held fixed throughout the simulation
- <b>Area of single platelet</b>: 500 Å<sup>2</sup> =~ 22.5 Å x 22.5 Å 
- <b>Distance</b> between platelets(H): 9 ,15, 30, 60 Å spacings are considered in the simulations, to study how different spacing influence the absorption process
- <b>Periodic boundary conditions (PBCs)</b> are applied in the x, y and z directions.


Other parameters for the simulation are:
- The H<sub>2</sub> molecules are approximated as <u>rigid spheres</u> with a mass double that of a hydrogen atom
- <b>Temperature</b>: 298 k,
- <b>Pressure</b>: 112 atm for the gas phase,
- <b>Simulation box sizes</b>: 60 Å (x), 60 Å (y), 36 Å (z, for H=9 Å: different sizes along the z axis are considered for different distances, to ensure correct PBCs)
</span>


### <span style="color:coral">2.2</span>. <b>Potentials</b>

1. <span style="color:coral"><b>Adsorbate-adsorbate interactions</b></span>: 
    The H<sub>2</sub>-H<sub>2</sub> interactions are modeled using the Silvera-Goldman <a href="#r5">[5]</a> potential:
    $$
    V_{Silvera-Goldman}(r) = \exp\left(\alpha - \beta r - \gamma r^2\right) - \left(\frac{C_{6}}{r^6} + \frac{C_{8}}{r^8} + \frac{C_{10}}{r^10}\right)f_{c}(r) + \frac{C_{9}}{r^9}f_{c}(r)
    $$(SG potential)
    where $r$ is the distance between the H<sub>2</sub> molecules, and $\alpha, \beta, \gamma, C_{6}, C_{8}, C_{10}$ are the potential parameters. 
    The fist term in the potential is the <u>exponential short-range repulsive term</u>, while the second term represents the <u>asymptotic van der Waals attraction, attenuated by $f_{c}(r)$ at short distances</u>.
    The purpose of $f_{c}(r)$  is to join the short- and long-range asymptotic behavior, cutting off the multipole terms which become infinite at the origin:

    $$
    f_{c}(r) = \left\{
        \begin{array}{ll}
            \exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] & \text{if } r \leq r_{c} \\
            1 & \text{otherwise}
        \end{array} \right.
    $$
    where $r_{c}$ is a parameter dependent on the position of the potential well minimum.

    Since the Silvera-Goldman potential is not implemented in LAMMPS, an extentsion of its C++ codebase is required: the implementation can be found in the <a href="/addingHydrogen/additionalPairs/pair_sg.cpp"><code>pair_sg.cpp</code></a> file. To do so, it is first necessary to compute the gradient of the potential, to derive the force acting on the particles used in the simulation.

    $$\vec{F}_{Silvera-Goldman}= -\nabla V_{Silvera-Goldman}(r) = \begin{bmatrix} F_{x} \\ F_{y} \\ F_{z} \end{bmatrix}$$

    where $F_{x}, F_{y}, F_{z}$ are the components of the force acting on the particles in the x, y, and z directions, respectively. Since the potential is a function of the distance between the particles, we will just write the force component along the x direction: the other components can be derived in an analogous way.

    $$
    F_{x} = \left\{
        \begin{array}{ll}
            \exp\left(\alpha - \beta r - \gamma r^2\right)\left(\frac{\beta x}{r} + 2\gamma x\right) - \left(\frac{6C_{6}x}{r^8} + \frac{8C_{8}x}{r^{10}} + \frac{10C_{10}x}{r^{12}} -\frac{9C_{9}x}{r^{11}}\right)  \qquad \text{ if } r > r_{c} \\
            \\
            \exp\left(\alpha - \beta r - \gamma r^2\right)\left(\frac{\beta x}{r} + 2\gamma x\right) - \left[\frac{6C_{6}x}{r^8} + \frac{8C_{8}x}{r^{10}} + \frac{10C_{10}x}{r^{12}} - \frac{9C_{9}x}{r^{11}}\right]\exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] - \\
             \left[\frac{C_{6}x}{r^6} + \frac{C_{8}x}{r^8} + \frac{C_{10}x}{r^{10}} - \frac{C_{9}x}{r^{9}}\right]\exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] \left(\frac{r_{c}}{r} - 1\right)\frac{2r_{c}x}{r^3} \qquad \text{otherwise}
        \end{array} \right.
    $$
    where $x$ is the distance between the particles in the x direction.

    This is the actual force implemented in the <code>pair_sg.cpp</code> file, which is then used in the LAMMPS simulation.
    
2. <span style="color:coral"><b>Adsorbate-adsorbent interactions</b></span>: 
    The graphite-H<sub>2</sub> interactions are modeled using the Crowell-Brown potential, as:

    $$
    V_{Crowell-Brown}(r) = 4 \epsilon_{CB} \left[\left(\frac{\sigma_{sf}}{r}\right)^{12} - \left(\frac{\sigma_{sf}}{r}\right)^{6}\frac{3\left(P_{\parallel}-P_{\perp}\right)\cos^2(\theta)+\left(P_{\parallel}+5P_{\perp}\right)}{4P_{\parallel}+2P_{\perp}}\right]
    $$

    where $r$ is the distance from the hydrogen molecule to the carbon atom, $\theta$ is the angle between the axis normal to the graphite sheet and a line connecting the hydrogen and carbon atoms, $\sigma_{sf}$ is the solid-fluid Lennard-Jones size parameter, $\sigma_{ff}$ is the fluid-fluid Lennard-Jones size parameter, and $\epsilon_{CB}$ is the well depth at $\theta=0$ written in terms of atomic parameters as:

    $$
        \epsilon_{CB} = \frac{E_{H}E_{C}P_{H}\left(4P_{\parallel}+2P_{\perp}\right)}{\left(E_{H}+E_{C}\right)\sigma_{ff}^6}
    $$

    the parameters are given in <a href="#r4">[4]</a> 

    In particular, this potential is chosen for its ability to account for the difference in polarizability of the graphene sheet parallel and perpendicular to the graphene plane. In this study, diffently from <a href="#r3">[3]</a>, the CB potential is not modified to make it longer-ranged. Even if at this level of study there would not be subtantial change, the reason for this choise is that the focus of that study was to implement a model that could be used to replicate the results of <a href="#r2">[2]</a>, results which later proved to be controversial.

    What makes this potential trickier to use with respect to the Silvera-Goldman is the dependence on the angle $\theta$. Nonetheless, it is possible to rewrite the potential taking into consideration that:

    $$
    \cos\theta = \frac{z}{r}
    $$

    where $z$ is the distance between the carbon atom and the graphite atom. This allows to rewrite the potential as a function of the distance between the carbon atom and the graphite plane, which is a more convenient quantity to use in the simulation.

    The force acting on the carbon atom can be derived from the potential, and is given by:

    $$
    \vec{F}_{Crowell-Brown} = \left\{
            \begin{array}{ll}
                -4{\epsilon}_{CB} \left(\frac{6{\sigma}_{CB}^{6} x \left(\frac{3 \left(P_{\parallel} - P_{\perp}\right) z^{2}}{x^{2} + z^{2} + y^{2}} + 5P_{\perp} + P_{\parallel}\right)}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{4}} + \frac{6 \left(P_{\parallel} - P_{\perp}\right) z^{2} {\sigma}_{CB}^{6} x}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{5}} - \frac{12{\sigma}_{CB}^{12} x}{\left(x^{2} + z^{2} + y^{2}\right)^{7}}\right)\\
                \\
                -4{\epsilon}_{CB} \left(\frac{6{\sigma}_{CB}^{6} y \left(\frac{3 \left(P_{\parallel} - P_{\perp}\right) z^{2}}{x^{2} + z^{2} + y^{2}} + 5P_{\perp} + P_{\parallel}\right)}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{4}} + \frac{6 \left(P_{\parallel} - P_{\perp}\right) z^{2} {\sigma}_{CB}^{6} y}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{5}} - \frac{12{\sigma}_{CB}^{12} y}{\left(x^{2} + z^{2} + y^{2}\right)^{7}}\right)\\
                \\
                -4{\epsilon}_{CB} \left(\frac{6{\sigma}_{CB}^{6} z \left(\frac{3 \left(P_{\parallel} - P_{\perp}\right) z^{2}}{x^{2} + z^{2} + y^{2}} + 5P_{\perp} + P_{\parallel}\right)}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{4}} - \frac{{\sigma}_{CB}^{6} \left(\frac{6 \left(P_{\parallel} - P_{\perp}\right) z}{x^{2} + z^{2} + y^{2}} - \frac{6 \left(P_{\parallel} - P_{\perp}\right) z^{3}}{\left(x^{2} + z^{2} + y^{2}\right)^{2}}\right)}{\left(2P_{\perp} + 4P_{\parallel}\right) \left(x^{2} + z^{2} + y^{2}\right)^{3}} - \frac{12{\sigma}_{CB}^{12} z}{\left(x^{2} + z^{2} + y^{2}\right)^{7}}\right)
            \end{array} \right.
        $$

    <div class="alert alert-block alert-warning"> <span style="font-size: 1.2em;"><b>&#x26A0 Observation </b>: </span>
    </br>
    It is important to notice that, normally, one should also account for quantum effects during the GCMC simulations, for instance, by integration of path integrals. Nonetheless, as highlighted in <a href="#r3">[3]</a>, at the considered temperatures and pressure, the quantum effects are expected to be negligible.
    </div>

    ---
    Since none of the two potentials are implemented in LAMMPS, new files for the <code>pairs</code> implementation in LAMMPS were required to performed the simulations. Those can be found in the <code>MD/additionalPairs</code>folder.

    

<span style="font-size: 1.1em;">
To asses the correctness of the potential profile used to model the interactions between the H<sub>2</sub> molecules and the graphite platelets, the adsorption barrier near an interface was measured with means of <b>Umbrella Sampling</b> simulations, togheter with a handy implementation of the <a href="http://membrane.urmc.rochester.edu/?page_id=126"> WHAM (Weighted Histogram Analysis Method) algorithm.</a>

<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>:
Umbrella sampling is a technique used in molecular simulations to efficiently sample systems with high energy barriers. It applies a biasing potential (often harmonic) to encourage exploration of otherwise hard-to-reach regions of configuration space. The system is divided into several windows along a reaction coordinate, with each simulation focusing on a specific region. Afterward, the results are combined using methods like WHAM to remove the bias and recover an unbiased free energy profile. It is particularly useful for calculating free energy differences and studying rare events in molecular systems.
</div>

</span>

```python
import math

PATH = os.getcwd()
PATH = PATH + '/MD/addingHydrogen/simulation_data'

def pot_profile_CB():
    # Evaluate the CB potential considering a single interaction C-H2
    x=np.linspace(0.05,15,100)
    sigma_CC=3.43 # Å
    sigma_HH=3.075 # Å
    sigma_CH_geo=pow(sigma_CC*sigma_HH,1/2) # Å
    sigma_CH_arit=0.5*(sigma_CC+sigma_HH) # Å 

    P_par=0.57 # Å^3
    P_perp=1.995 # Å^3
    P_h=0.81 # Å^3
    E_h=2.337e5*(1/503.23) # K * 1/503.23 kcal/mol/K = kcal/mol
    E_c=1.39254e5*(1/503.23) # K * 1/503.23 kcal/mol/K = kcal/mol

    epsilon_CB=4*E_h*E_c*P_h*(P_par+P_perp/2)/(pow(sigma_HH,6)*(E_h+E_c)) # kcal/mol 
    eps= epsilon_CB

  
    V_CB=eps*(pow(sigma_CH_geo/x,12)-pow(sigma_CH_geo/x,6)*((3*(P_par-P_perp)+(P_par+5*P_perp))/(4*P_par+2*P_perp)))
    
    # now add mirrored elements to the array
    x=np.linspace(-15,15,200)
    V_CB = np.concatenate((V_CB,V_CB[::-1]))

    min_V_CB = np.min(V_CB)

    fig, ax = plt.subplots(1,1,figsize=(10,6))
    ax.plot(x, V_CB, color = 'coral')
    ax.set_xlabel(r'$Z$ (Å)', fontdict=font)
    ax.set_ylabel(r'$\Phi_{CB}$ (kcal/mol)', fontdict=font)
    ax.set_title('Adsorption potential (Single contribution) - H = 30 Å', fontdict=font, fontsize=16)
    # Set grid minor ticks on y axis
    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.05))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.25))
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    ax.set_ylim(min_V_CB-0.2, -min_V_CB+0.2)


def plot_abpot(file_path):
    data = np.loadtxt(file_path)
    x = data[:,0]
    y = data[:,1]
    fig, ax = plt.subplots(1,1,figsize=(10,6))
    ax.plot(x, y, marker= "D", markeredgecolor='black', markerfacecolor='black', color = 'coral', markersize=3)
    ax.set_xlabel(r'$Z$ (Å)', fontdict=font)
    ax.set_ylabel(r'$\Phi_{tot}$ (kcal/mol)', fontdict=font)
    ax.set_title('Adsorption potential (Total contribution) - H = 30 Å', fontdict=font, fontsize=16)
    # Set grid minor ticks on y axis
    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.25))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.25))
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')


pot_profile_CB()
plot_abpot(PATH + '/PMF.dat')
```

```python
# Plotting hydrogen number increase over the simulation:

hydrogen_trend_directory = PATH + '/hydrogen_counts/eps=80eps_CB/'

def plot_hydrogen_increase(hydrogen_directory):
    fig, ax = plt.subplots(1,1,figsize=(10,6))
    #change the colormap for the lines
    ax.set_prop_cycle(color=plt.cm.viridis(np.linspace(0.,0.8,4)))
    # For each file in the directory, open it, skip the first 2 lines and read the number of hydrogen atoms
    start_times = [90000, 80000, 70000, 60000]
    for file in os.listdir(hydrogen_directory):
        with open(hydrogen_directory+file, 'r') as f:
            for i in range(2):
                f.readline()
            data = np.loadtxt(f)
            time = data[:,0]
            hydrogen_number = data[:,1]
            # The files are titled as: 'hydrogenCount9.data', 'hydrogenCount15.data', etc.,
            H = file.split('.')[0].split('t')[-1]
            if H not in ['9', '15', '30', '60']:
                    raise ValueError('The hydrogen atom distance is not valid')
            # Plot the number of hydrogen atoms over time and save the color for the mean value line
            ax.plot(time, hydrogen_number, label='H = '+H+' Å')
            color = plt.gca().lines[-1].get_color()
            # Find the mean values for the values after time = 10000 ps
            mean_value_after = np.mean(hydrogen_number[time>130000])
            #mean_value_before = np.mean(hydrogen_number[time<10000])
            times_after = time[time>start_times.pop()]
            ax.plot(times_after, np.ones(len(times_after))*mean_value_after, linestyle='--', color='black')
            # Enclose the text in a rounded box
            ax.text(times_after[0], mean_value_after+2, str(round(mean_value_after, 2)), fontsize=12, color='white', bbox=dict(facecolor='white', edgecolor=color, boxstyle='round,pad=0.5'))
            # Add the values right to the left of the lines
            ax.text(times_after[0], mean_value_after+2, str(round(mean_value_after, 2)), fontsize=12, color=color)
    ax.set_xlabel('Time (ps)', fontdict=font)
    ax.set_ylabel(r'$N_{atoms}$', fontdict=font)
    eps_directory = hydrogen_directory.split('/')
    eps_directory = eps_directory[-2]
    eps_directory = eps_directory.split('=')[-1]
    eps_mul = eps_directory.split('e')[0]
    ax.set_title('Number of hydrogen atoms over time: '+r'$\epsilon='+eps_mul+r'\epsilon_{CB}$', fontdict=font, fontsize=16)
        
    ax.legend()

plot_hydrogen_increase(hydrogen_trend_directory)
```

### <span style="color:coral">2.5</span>. <b>Excess weights for different value of $\epsilon$ </b>


```python
# Compute the excess adsorption:

def compute_excess_adsorption(hydrogen_directory):
    fig, ax = plt.subplots(1,1,figsize=(10,6))
    
    # Determine the number of atoms in the vessel without the absorbant
    for dimension in ['big', 'medium' ,'small']:
        with open(PATH+'/hydrogen_counts/hydrogenCount_'+dimension+'.data', 'r') as f:
            for i in range(2):
                f.readline()
            data = np.loadtxt(f)
            time = data[:,0]
            hydrogen_number = data[:,1]
            mean_value = np.mean(hydrogen_number[time>100000])
            if dimension == 'big':
                mean_value_big = mean_value
            elif dimension == 'medium':
                mean_value_medium = mean_value
            else:
                mean_value_small = mean_value

    if isinstance(hydrogen_directory, str):
        hydrogen_directory = [hydrogen_directory]

    colors = plt.cm.viridis(np.linspace(0.,0.8,len(hydrogen_directory)))
    colors = list(colors)
    for directory in hydrogen_directory:
        color = colors.pop()
        weight_fractions = []
        # For each file in the directory, open it, skip the first 2 lines and read the number of hydrogen atoms
        for file in os.listdir(directory):
            with open(directory+file, 'r') as f:
                for i in range(2):
                    f.readline()
                data = np.loadtxt(f)
                time = data[:,0]
                hydrogen_number = data[:,1]
                # The files are titled as: 'hydrogenCount9.data', 'hydrogenCount15.data', etc.,
                H = file.split('.')[0].split('t')[-1]
                if H not in ['9', '15', '30', '60']:
                    raise ValueError('The hydrogen atom distance is not valid')
                mean_value_after = np.mean(hydrogen_number[time>100000])
                times_after = time[time>80000]
                # Compute the excess adsorption
                hydrogen_excess = mean_value_after - (mean_value_big if H == '60' else mean_value_medium if (H == '30' or H== '15') else mean_value_small if H == '9' else 0)
                carbon_atoms = 960 if H == '9' else 960/2 if H == '15' else 960/4 if (H == '30' or H == '60') else 0
                hydrogen_weight = 2.01588
                carbon_weight = 12.011
                weight_fraction = (hydrogen_excess*hydrogen_weight)/(carbon_atoms*carbon_weight + hydrogen_excess*hydrogen_weight)
                # Append the weight fraction to the list
                weight_fractions.append(weight_fraction)
        eps_directory = directory.split('/')
        eps_directory = eps_directory[-2]
        eps_directory = eps_directory.split('=')[-1]
        eps_mul = eps_directory.split('e')[0]
        label=r'$\epsilon='+eps_mul+r'\epsilon_{CB}$'
        ax.plot([9, 15, 30, 60], weight_fractions, marker='D', markeredgecolor='black', markerfacecolor='black', color=color, label=label)

    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.001))
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    ax.set_xlabel('H (Å)', fontdict=font)
    ax.set_ylabel('Weight fractions (%)', fontdict=font)
    ax.set_title(r'Excess adsorption for different $\epsilon$ and H values', fontdict=font, fontsize=16)
    ax.legend()
    # Set grid minor ticks on y axis    

compute_excess_adsorption([PATH + '/hydrogen_counts/eps=eps_CB/', PATH + '/hydrogen_counts/eps=4eps_CB/', PATH + '/hydrogen_counts/eps=20eps_CB/', PATH + '/hydrogen_counts/eps=80eps_CB/'])
    
```

<span style="font-size: 1.1em;">

The values of <b>weight excess</b> are relatively close to the one reported on <a href="#r3">[3]</a>, with the exception of the ones for $H=15$. Namely, the authors reported a value of $0.46%$, (which is a factor of 100 less than that reported experimentally in <a href="#r2">[2]</a>), that is not too far from the one of approximately $0.4%$ reported for the same value of $\epsilon$, at the same temperature and pressure values. It is also to take into account the fact that, in this study, the planes of graphene have not been considered rigidly in their original positions.
As emerges from <a href="#r1">[1]</a>, a <i>good enough</i> parametrization for the interaction between hydrogen and carbon lies in the interval $\left[\epsilon_{CB}, 4\epsilon_{CB}\right]$, while higher values where tried to fit the experimental data obtained by Rodriguez et al. in <a href="#r2">[2]</a>. 
</br>

Additionally, as noted by Wang et al. in <a href="#r3">[3]</a>, such experimental results are not consistent with any reasonable physisorption model: assuming that the solid-fluid potential is not very seriously in error, the results indicate that no slit pore geometries are capable of adsorbing
the amount of H reported by Rodriguez et al., which would only be possible considering $\epsilon$ in a neighborhood of $150\epsilon$. Unrealistically strong solid-fluid potentials are needed to reproduce experimental data, even considering long-ranged potential. This may find an explanation if we start considering chemisorption phenomena as initially indicated Rodriguezetal, as well as the presence of metals capable of dissociatively absorbing hydrogen, dispersed in the nanofibers as a result of the process used to remove the metal catalyst particles, employed to produce the GNFs.

</span>
