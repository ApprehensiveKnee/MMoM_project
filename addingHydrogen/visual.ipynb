## <span style="color:coral">0</span>. <b>Introduction</b>

<span style="font-size: 1.1em;">
The adsorption of gases on carbon nanofibers (CNFs) has garnered significant interest due to its wide-ranging applications in fields such as <strong>environmental remediation</strong>, <strong>gas storage</strong>, and <strong>catalysis</strong>. Carbon nanofibers, characterized by their high surface area, tunable porosity, and unique electronic properties, present an ideal material for gas adsorption processes.

Understanding the adsorption mechanisms and capacities of gases on CNFs is crucial for optimizing these processes and enhancing the performance of related technologies.
<br>

<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>: <u>Absorption types: physisorption and chemisorption</u> <br>
Physisorption and chemisorption are two types of physical adsorption, which refer to extracting a substance from a medium by interacting with solid surfaces:
<ul>
    <li><b>Physisorption</b> is the adsorption in which the molecular interactions between the adsorbate molecules and the adsorbent are primarily <b>governed by van der Waals forces</b>;
    <li><b>Chemisorption</b> is the adsorption in which the molecular interactions between the adsorbate molecules and the adsorbent are primarily governed by <b>chemical bonds</b>.
</ul>
In this study, we will <b>focus specifically on the physisorption of gases on CNFs</b>.
In general, Physisorption is faster and more effective than chemisorption. Additionally, Physisorption is more selective, meaning that it can selectively adsorb certain substances over others.
None the less, physisorption is non-specific in nature and is generally regarded as a weak, reversible process governed by competitive adsorption and desorption</div>

</span>

<span style="font-size: 1.1em;">
In our study, we always consider <b>platelet GNFs</b>: the pores are formed between the graphite platelets, and the adsorption sites are located on the platelet surfaces. The areas of the platelets, as well as the distance between them, can be tuned to optimize the adsorption capacity and selectivity of the CNFs.
</span>

<span style="font-size: 1.1em;">
<div class="alert alert-block alert-info"> <b>&#x24D8 Note </b>: <u>GNF structures</u> <br>
GNFs consist of parallel stacks of graphite platelets that are oriented parallel, perpendicular, or at an angle with respect to the fiber axis. These structures are termed <b>tubular</b>, <b>platelet</b>, and <b>herringbone</b>, respectively
</div>
</span>

## <span style="color:coral">1</span>. <b>DFT analysis</b>

### <span style="color:coral">1.1</span>. <b>About the typical strenght of the interaction...</b>
<span style="font-size: 1.1em;">
As highlighted by Morris et al. (2007) <a href="#r1">[1]</a>, typically, the binding energy must be on the order of <b>25 kJ/mol (0.25 eV/molecule)</b> or higher to have significant storage at operating temperatures and pressures, while not irreversibly binding the hydrogen. The typical sheet spacing for carbon layers is 3.35 Å, which does not allow for significant hydrogen storage, in spite of the hydrogen weak attraction to the carbon surface. 
Nonetheless, the absorption/storage can be significantly enhanced by increasing the interlayer spacing of the graphite platelets.
<br>

In this regard, one method - through which the GNFs considered in this study are generated<a href="#r2">[2]</a> - consists in <b>catalyzed decomposition of carbon-containing gases and their mixtures over selected metals and alloys at temperatures between 450 and 750°C</b>. IN the various GNFs conformations, the layers are separated at distances that are dependent upon the nature of the catalyst and the gas phase, as well as the reaction conditions, where the minimum value possible is that of single-crystal graphite, 3.35 Å.
</span>

### <span style="color:coral">2.2</span>. <b>Potentials</b>

1. <span style="color:coral"><b>Adsorbate-adsorbate interactions</b></span>: 
    The H<sub>2</sub>-H<sub>2</sub> interactions are modeled using the Silvera-Goldman potential:
    $$
    V_{Silvera-Goldman}(r) = \exp\left(\alpha - \beta r - \gamma r^2\right) - \left(\frac{C_{6}}{r^6} + \frac{C_{8}}{r^8} + \frac{C_{10}}{r^10}\right)f_{c}(r) + \frac{C_{9}}{r^9}f_{c}(r)
    $$(SG potential)
    where $r$ is the distance between the H<sub>2</sub> molecules, and $\alpha, \beta, \gamma, C_{6}, C_{8}, C_{10}$ are the potential parameters. 
    The fist term in the potential is the <u>exponential short-range repulsive term</u>, while the second term represents the <u>asymptotic van der Waals attraction, attenuated by $f_{c}(r)$ at short distances</u>.
    The purpose of $f_{c}(r)$  is to join the short- and long-range asymptotic behavior, cutting off the multipole terms which become infinite at the origin:

    $$
    f_{c}(r) = \left\{
        \begin{array}{ll}
            \exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] & \text{if } r \leq r_{c} \\
            1 & \text{otherwise}
        \end{array} \right.
    $$
    where $r_{c}$ is a parameter dependent on the position of the potential well minimum.

    Since the Silvera-Goldman potential is not implemented in LAMMPS, an extentsion of its C++ codebase is required: the implementation can be found in the <a href="/addingHydrogen/additionalPairs/pair_sg.cpp"><code>pair_sg.cpp</code></a> file. To do so, it is first necessary to compute the gradient of the potential, to derive the force acting on the particles used in the simulation.

    $$\vec{F}_{Silvera-Goldman}= -\nabla V_{Silvera-Goldman}(r) = \begin{bmatrix} F_{x} \\ F_{y} \\ F_{z} \end{bmatrix}$$

    where $F_{x}, F_{y}, F_{z}$ are the components of the force acting on the particles in the x, y, and z directions, respectively. Since the potential is a function of the distance between the particles, we will just write the force component along the x direction: the other components can be derived in an analogous way.

    $$
    F_{x} = \left\{
        \begin{array}{ll}
            \exp\left(\alpha - \beta r - \gamma r^2\right)\left(\frac{\beta x}{r} + 2\gamma x\right) - \left(\frac{6C_{6}x}{r^8} + \frac{8C_{8}x}{r^{10}} + \frac{10C_{10}x}{r^{12}} -\frac{9C_{9}x}{r^{11}}\right)  \qquad \text{ if } r > r_{c} \\
            \\
            \exp\left(\alpha - \beta r - \gamma r^2\right)\left(\frac{\beta x}{r} + 2\gamma x\right) - \left[\frac{6C_{6}x}{r^8} + \frac{8C_{8}x}{r^{10}} + \frac{10C_{10}x}{r^{12}} - \frac{9C_{9}x}{r^{11}}\right]\exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] - \\ \left[\frac{C_{6}x}{r^6} + \frac{C_{8}x}{r^8} + \frac{C_{10}x}{r^{10}} - \frac{C_{9}x}{r^{9}}\right]\exp\left[-\left(\frac{r_{c}}{r} - 1\right)^2\right] \left(\frac{r_{c}}{r} - 1\right)\frac{2r_{c}x}{r^3} \qquad \text{otherwise}
        \end{array} \right.
    $$
    where $x$ is the distance between the particles in the x direction.

    This is the actual force implemented in the <code>pair_sg.cpp</code> file, which is then used in the LAMMPS simulation.
